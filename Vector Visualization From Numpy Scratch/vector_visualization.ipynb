```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-3, 3, 10)   # 10 points from -3 to 3
y = np.linspace(-3, 3, 10)
X, Y = np.meshgrid(x, y)

# Step 3: Combine into coordinate pairs
points = np.vstack([X.ravel(), Y.ravel()])

# Step 4: Plot the grid of points
plt.scatter(points[0], points[1], color='gray', s=40)#s = point size
plt.title("2D Grid of Points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid(True)   # optional grid lines
plt.gca().set_aspect('equal')  # keep square aspect ratio
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create grid
x = np.linspace(-3, 3, 10)
y = np.linspace(-3, 3, 10)
X, Y = np.meshgrid(x, y)
points = np.vstack([X.ravel(), Y.ravel()])

# Step 2: Define transformation matrices
scale = np.array([[1.5, 0], [0, 0.5]])
theta = np.pi / 4
rotation = np.array([[np.cos(theta), -np.sin(theta)],
                     [np.sin(theta),  np.cos(theta)]])
shear = np.array([[1, 0.5], [0, 1]])
reflect = np.array([[1, 0], [0, -1]])

# Step 3: Apply transformations
scaled = np.dot(scale, points)
rotated = np.dot(rotation, points)
sheared = np.dot(shear, points)
reflected = np.dot(reflect, points)

# Step 4: Define helper function for repeated styling
def show_plot(transformed_points, title, label):
    plt.scatter(points[0], points[1], color='gray', alpha=0.5, label='Original')
    plt.scatter(transformed_points[0], transformed_points[1], color='blue', label=label)
    plt.title(title)
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.legend()
    plt.grid(True)
    plt.gca().set_aspect('equal')
    plt.show()
```

```python

#  Rotation
show_plot(rotated, "Rotation Transformation", "Rotated")
```

```python
# 4 Reflection
show_plot(reflected, "Reflection Transformation", "Reflected")
```
