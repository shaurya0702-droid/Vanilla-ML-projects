```python
X_train = train[['Time_Period']].values.astype(float) # double brackets to keep 2d shape 
y_train = train['Value'].values.astype(float)
X_test = test[['Time_Period']].values.astype(float)
y_test = test['Value'].values.astype(float)
# Extract features and convert to numpy arrays 
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
# Feature scaling
# Features often have different units and magnitudes. For example, age might range 0-100, income could be in thousands or lakhs.
# Models that use gradient descent, distance calculations, or regularization (e.g., linear regression, logistic regression, SVM, k-NN) perform better and train faster if the input features are on a similar scale.
# Without scaling, features with larger ranges can dominate the learning process or cause numerical instability.

X_mean = X_train.mean(axis=0)  # Calculates the mean and standard deviation
X_std = X_train.std(axis=0)  # axis=0 means for each feature => column
X_train_scaled = (X_train - X_mean) / X_std
X_test_scaled = (X_test - X_mean) / X_std

# This is z score normalization mean becomes 0 std dev becomes 1

#Why Standardization Was Needed (Even With 1 Feature)
#without it Large values caused NaN: Raw feature values were too large, causing overflow in calculations.
#Learning rate couldn't handle it: Fixed learning rate 0.01 worked poorly with large-scale data, causing unstable updates.
#Scaling fixed instability: Converting to mean 0, std 1 made gradients manageable and training stable.
```

```python
#Model definition with loss tracking
class SimpleLinearRegression:
    def __init__(self):
        self.w = 0.0
        self.b = 0.0
        self.losses = []
    def fit(self, X, y, iterations=1000, learning_rate=0.01): #iteration =epochs
        n = len(y)
        X_flat = X.flatten()#Flattens X from(n,1) to  (n,)-1d array  for simpler calculations
        for i in range(iterations):
            y_pred = self.w * X_flat + self.b
            error = y_pred - y
            loss = np.mean(error ** 2)
            self.losses.append(loss)#Records the loss value for this iteration into the losses list for tracking.

            #gradient descent update
            self.w -= (learning_rate * 2/n) * np.sum(X_flat * error) #how much the slope w is responsible for the error overall.
            self.b -= (learning_rate * 2/n) * np.sum(error) #how much the bias b is responsible for the error overall.
            
    def predict(self, X):
        return self.w * X.flatten() + self.b
    
    def mse(self, y_true, y_pred):
        return np.mean((y_true - y_pred) ** 2)
    def rmse(self, y_true, y_pred):
        return np.sqrt(self.mse(y_true, y_pred))
    def r2_score(self, y_true, y_pred): #R score (coefficient of determination) tells how well your regression line explains the variation in the target y.
        ss_res = np.sum((y_true - y_pred) ** 2)
        ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
        return 1 - (ss_res / ss_tot)
```

```python
# Train the model
model = SimpleLinearRegression()
model.fit(X_train_scaled, y_train, iterations=1000, learning_rate=0.01)
```

```python
#Evaluate 
y_pred = model.predict(X_test_scaled)
print(f"MSE: {model.mse(y_test, y_pred):.6f}")
print(f"RMSE: {model.rmse(y_test, y_pred):.6f}")
print(f"R Score: {model.r2_score(y_test, y_pred):.6f}")
print(f"Accuracy: {model.r2_score(y_test, y_pred)*100:.2f}%")
```
